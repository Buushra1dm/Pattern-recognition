```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

**step 2**

```python
num_hours_studied = np.array([1, 3, 3, 4, 5, 6, 7, 7, 8, 8, 10])
exam_score = np.array([18, 26, 31, 40, 55, 62, 71, 70, 75, 85, 97])
```

**step 3**

```python
plt.scatter(num_hours_studied, exam_score)
plt.xlabel('Number of hours studied', fontsize=12)
plt.ylabel('Exam Score', fontsize=12)
plt.title('Exam Score & Hours Studied')
plt.show()
```

**step 4**

```python
X = np.column_stack((np.ones_like(num_hours_studied), num_hours_studied))
theta_best = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(exam_score)
```

```python
theta_best
```

**step 5**

```python
plt.scatter(num_hours_studied, exam_score)
plt.plot(num_hours_studied, X.dot(theta_best), color='red', label='Reg Line')
plt.xlabel('Number of hours studied',fontsize=12)
plt.ylabel('Exam Score',fontsize=12)
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
x_pred = 9
y_predicted = np.dot([1, x_pred], theta_best)
print(f"The predicted value of x is 9: {y_predicted:.3f}")
```

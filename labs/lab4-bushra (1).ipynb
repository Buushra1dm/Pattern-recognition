```python
# Split the dataset into X and y
X = df.drop('booking_status', axis=1)
y = df['booking_status']
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
# Create a decision tree classifier and fit it to the training data
model = DecisionTreeClassifier(random_state=100)
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy is:", accuracy)
```

- There is an error due to categorical columns in the dataset
- we will usr use numeric datatypes insted

```python
# Select only the numeric features
X_numeric = X.select_dtypes(exclude=['object'])
```

```python
# Split the dataset into training and testing sets
X_train_numeric, X_test_numeric, y_train, y_test = train_test_split(X_numeric, y, test_size=0.2, random_state=100)

# Create a decision tree classifier and fit it to the training data
clf_numeric = DecisionTreeClassifier(random_state=100)
clf_numeric.fit(X_train_numeric, y_train)

# Make predictions on the testing data
y_pred_numeric = clf_numeric.predict(X_test_numeric)

# Evaluate the accuracy of the model
accuracy_numeric = accuracy_score(y_test, y_pred_numeric)
print("Accuracy (numeric features only):", accuracy_numeric)
```

This Accuracy is %85 accurate which is good

```python
from sklearn.model_selection import GridSearchCV


# Define the hyperparameter grid
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [5, 10, 15, None],
    'max_features': ['sqrt', 'log2', None],
    'splitter': ['best', 'random']
}

# Create a decision tree classifier
clf_gs = DecisionTreeClassifier(random_state=100)

# Use GridSearchCV to find the best hyperparameters
clf_gs = GridSearchCV(clf_gs, param_grid, cv=5)
clf_gs.fit(X_train_numeric, y_train)

# Print the best hyperparameters
print("Best hyperparameters:", clf_gs.best_params_)
```

```python
# Create a decision tree classifier with the best hyperparameters
clf_best = DecisionTreeClassifier(criterion ='gini', max_depth = 15, max_features =None, splitter ='best',random_state=100)
clf_best.fit(X_train_numeric, y_train)

# Make predictions on the testing data
y_pred_best = clf_best.predict(X_test_numeric)

# Evaluate the accuracy of the model
accuracy_best = accuracy_score(y_test, y_pred_best)
print("Accuracy (numeric features only, with best hyperparameters):", accuracy_best)
```

```python
print("Accuracy (numeric features only, without tuning hyperparameters):", accuracy_numeric)
print("Accuracy (numeric features only, with best hyperparameters):", accuracy_best)
```

- Accuracy is almost same . improved a litlle

```python
column_transformer = make_column_transformer((OneHotEncoder(), ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']),remainder='passthrough')

# First, we must transform the entire dataset
X_transformed = column_transformer.fit_transform(X)
X_transformed = pd.DataFrame(data=X_transformed)

#second,we divide the transformed dataset into train/test dataset
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, random_state = 100)
```

```python
# Encode categorical features
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type'])], remainder='passthrough')
X = ct.fit_transform(df.drop('booking_status', axis=1))

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

# Create a decision tree classifier
clf_all = DecisionTreeClassifier(random_state=100)

# Fit the classifier to the training data
clf_all.fit(X_train, y_train)

# Make predictions on the testing data
y_pred_all = clf_all.predict(X_test)

# Evaluate the model
acc_all = accuracy_score(y_test, y_pred_all)
print("Accuracy (all features):", acc_all)
```

This model is %86.91 acuurate which means is better than the numeric only model and almost as good as the best-hyperparamter model

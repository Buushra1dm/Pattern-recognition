```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_hours_studied = np.array([1, 3, 3, 4, 5, 6, 7, 7, 8, 8, 10])
exam_score = np.array([18, 26, 31, 40, 55, 62, 71, 70, 75, 85, 97])
```

```python
plt.scatter(num_hours_studied, exam_score)
plt.xlabel('Number of hours studied', fontsize=12)
plt.ylabel('Exam Score', fontsize=12)
plt.title('Exam Score & Number of hours studied')
plt.show()
```

```python
#function that claculate batch gradient descent
def gradient_descent(X, y, l_rate=0.1, iteration=100):
    m = len(y)
    theta = np.zeros(2)
    X_b = np.c_[np.ones((m, 1)), X]
    #for loop to 100
    for _ in range(iteration):
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta -= l_rate * gradients
        
    return theta
```

```python
X = num_hours_studied.reshape(-1, 1)
y = exam_score
```

```python
theta = gradient_descent(X, y)


plt.scatter(X, y)
plt.plot(X, np.dot(np.c_[np.ones((len(X), 1)), X], theta), color='red')
plt.xlabel('Number of hours studied', fontsize=12)
plt.ylabel('Exam Score', fontsize=12)
plt.title('Batch Gradient Descent')
plt.show()

print("learned theta is :", theta)
```

```python
learning_rate = [1, 2]

for lr in learning_rate:
    theta = gradient_descent(X, y, l_rate=lr)
    
    plt.scatter(X, y)
    plt.plot(X, np.dot(np.c_[np.ones((len(X), 1)), X], theta), color='red')
    plt.xlabel('Number of hours studied', fontsize=12)
    plt.ylabel('Exam Score', fontsize=12)
    plt.title(f'Learning Rate {lr}')
    plt.show()
    
    print(f"Learned theta for learning rate {lr}:", theta)
```

```python
#two different learning rates 
learning_rate = [0.02, 0.5]

for lr in learning_rate:
    theta = gradient_descent(X, y, l_rate=lr)
    
    plt.scatter(X, y)
    plt.plot(X, np.dot(np.c_[np.ones((len(X), 1)), X], theta), color='red')
    plt.xlabel('Number of hours studied', fontsize=12)
    plt.ylabel('Exam Score', fontsize=12)
    plt.title(f'Learning Rate {lr}')
    plt.show()
    
    print(f"Learned theta for learning rate {lr}:", theta)
```

different learning rates with fixed iteration when learning rates is small the learned theta come bigger
